```python
import folium

# Create Kolkata map
kolkata_map = folium.Map(
    location=[22.5726, 88.3639],  # Kolkata center
    zoom_start=12,
    tiles='OpenStreetMap'
)

# Add some known waterlogging spots
flood_spots = [
    {"name": "Park Circus", "loc": [22.5379, 88.3476], "severity": "high"},
    {"name": "Behala", "loc": [22.4868, 88.3085], "severity": "high"},
    {"name": "Lake Gardens", "loc": [22.5069, 88.3645], "severity": "medium"},
]

for spot in flood_spots:
    color = 'red' if spot['severity'] == 'high' else 'orange'
    folium.Marker(
        spot['loc'],
        popup=f"{spot['name']}: {spot['severity']} risk",
        icon=folium.Icon(color=color)
    ).add_to(kolkata_map)

# Save the map
kolkata_map.save('my_first_flood_map.html')
print("ðŸŽ‰ Map created! Check my_first_flood_map.html")

kolkata_map  # Display in notebook
```
