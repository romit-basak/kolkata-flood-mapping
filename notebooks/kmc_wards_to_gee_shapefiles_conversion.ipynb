```python
import geopandas as gpd
from pathlib import Path

DATA_DIR = Path('../data')
WARDS_DIR = DATA_DIR / 'wards'

# Load GeoJSON
wards = gpd.read_file(WARDS_DIR / 'kmc_wards_gee_ready.geojson')

# Save as Shapefile (creates .shp, .shx, .dbf, .prj files)
wards.to_file(WARDS_DIR / 'kmc_wards_gee_ready.shp', driver='ESRI Shapefile')

print("✓ Converted to Shapefile format")
print(f"✓ Files created in {WARDS_DIR}:")
print("   • kmc_wards_gee_ready.shp")
print("   • kmc_wards_gee_ready.shx")
print("   • kmc_wards_gee_ready.dbf")
print("   • kmc_wards_gee_ready.prj")
```
